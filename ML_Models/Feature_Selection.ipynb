```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
from sklearn.datasets import load_iris
```

```python
data=load_iris()
```

```python
data
```

```python
iris= pd.DataFrame(data=data.data, columns=data.feature_names)
```

```python
iris
```

```python
iris['label']=data.target
```

```python
iris
```

```python
x=np.array(iris.drop(['label'], axis=1))
```

```python
x
```

```python
y=np.array(iris['label'])
```

```python
y
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
sel= VarianceThreshold(threshold=0.6)
```

```python
vt_x=sel.fit_transform(x)
```

```python
vt_x
```

```python
df_vt_x= pd.DataFrame(vt_x)
```

```python
df_vt_x
```

```python
sns.scatterplot(x=0,y=1,data=df_vt_x, hue=y)
```
