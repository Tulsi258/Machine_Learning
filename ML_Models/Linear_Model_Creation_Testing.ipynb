```python
data_no_multicolinearity = new_data.drop(['Year'],axis=1)
```

```python
# deal Categorical variables
```

```python
data_dummies = pd.get_dummies(data_no_multicolinearity, drop_first=True)
```

```python
data_dummies
```

```python
inputs= data_dummies.drop(['Log_Price'], axis=1) #x
targets= data_dummies['Log_Price'] #y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler= StandardScaler()
```

```python
scaler.fit(inputs)
```

```python
input_scaled = scaler.transform(inputs)
```

```python
input_scaled
```

```python
input_scaled.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(input_scaled, targets, test_size=0.2, random_state=0)
```
