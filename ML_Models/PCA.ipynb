```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
sns.set_style('whitegrid')
```

```python
dataset = load_iris()
```

```python
dataset
```

```python
iris = pd.DataFrame(data=dataset.data,columns=dataset.feature_names)
```

```python
iris['label']=dataset.target
```

```python
iris
```

```python
 x=np.array(iris.drop(['label'], axis=1))
```

```python
x
```

```python
y=np.array(iris['label'])
```

```python
y
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
x_decomposed = pca.fit_transform(x)
```

```python
x_decomposed.shape
```

```python
x_decomposed
```

```python
x_decomposed = pd.DataFrame(x_decomposed, columns =['a','b'])
```

```python
x_decomposed
```

```python
sns.scatterplot(x='a',y='b', data=x_decomposed, hue=y)
```

```python
# Kernel PCA
```

```python
from sklearn.datasets import make_circles
from sklearn.decomposition import KernelPCA
```

```python
x,y = make_circles(n_samples=400, factor=0.3, noise=0.05)
```

```python
x
```

```python
y
```

```python
x_test = pd.DataFrame(x, columns=['a','b'])
```

```python
sns.scatterplot(x='a',y='b',data=x_test,hue=y)
```

```python
kpca = KernelPCA(kernel='rbf', gamma=10)
```

```python
x_LPCA= kpca.fit_transform(x)
```

```python
x_LPCA
```

```python
x_kpca=pd.DataFrame(x_LPCA)
```

```python
x_kpca
```

```python
fig,ax=plt.subplots(1,2,figsize=(8,6))
sns.scatterplot(x='a',y='b', data=x_test,ax=ax[0])
sns.scatterplot(x=0,y=1, data=x_kpca,ax=ax[1],hue=y)
```

```python
sns.scatterplot(x=0,y=1,hue=y,data=x_kpca)
```
